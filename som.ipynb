```python
from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt
from ddkdk import Som

from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_data = iris.data[:, :2]
iris_label = iris.target
```

```python
# https://pypi.org/project/sklearn-som/

x, y = np.meshgrid(np.arange(0.0, 3.0, 1.0), np.arange(0.0, 1.0, 1.0))
grid = np.vstack([x.ravel(), y.ravel()])
grid = grid.T

som = Som(2, 2)

som.fit(iris_data, grid, max_iter=100)

latent = som.latent
predictions = latent[:, 0]

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(5,7))
x = iris_data[:,0]
y = iris_data[:,1]
colors = ['red', 'green', 'blue']

ax[0].scatter(x, y, c=iris_label, cmap=ListedColormap(colors))
ax[0].title.set_text('Actual Classes')
ax[1].scatter(x, y, c=predictions, cmap=ListedColormap(colors))
ax[1].title.set_text('SOM Predictions')
plt.show()
```
